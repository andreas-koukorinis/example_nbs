```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sgmresearchbase.coint.regression import add_constant, ols, coch_regression
from sgmresearchbase.coint.simulation import gen_ar1
```

```python
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (16, 8)
```

```python
def sample_params():
    return  [0.0, 2*np.random.rand()-1]
```

```python
def simulate(l, params, alpha=0.99, sigma=10.0):
    x = np.arange(l)
    y = params[1]*x + params[0] + gen_ar1(l, alpha=alpha, sigma=sigma)

    return x, y
```

```python
def ols_model(x, y):
    return ols(add_constant(x), y).params
```

```python
def co_model(x, y):
    return coch_regression(add_constant(x), y).params
```

```python
def evaluate_model(m, n_evals=10000):
    all_diffs = []

    for i in range(n_evals):
        params = sample_params()
        x, y = simulate(500, params)
        est = m(x, y)
        all_diffs.append([params[0] - est[0], params[1] - est[1]])

    return np.array(all_diffs)
```

```python
for i in range(10):
    plt.plot(*simulate(500, sample_params()))
plt.title('Sample of time series (standard linear model with AR(1) errors).')
```

```python
diffs_ols = evaluate_model(ols_model)
diffs_co = evaluate_model(co_model)
```

```python
sns.distplot(diffs_ols[:, 0])
sns.distplot(diffs_co[:, 0])
plt.title('Disitribution of errors on the estimation of the intercept.')
```

```python
sns.distplot(diffs_ols[:, 1])
sns.distplot(diffs_co[:, 1])
plt.title('Disitribution of errors on the estimation of the gradient.')
```
