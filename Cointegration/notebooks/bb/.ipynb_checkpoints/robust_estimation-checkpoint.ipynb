```python
def get_basis(vals, model=1):
    if vals.shape[0] < 1:
        return np.zeros((2,))

    both_bases = CointegrationService.get_cointegrating_bases(vals, model=model)
    if both_bases is None:
        return np.zeros((2,))

    if both_bases[0, 0]*both_bases[1, 0] < 0:
        basis = both_bases[:, 0]
    elif both_bases[0, 1]*both_bases[1, 1] < 0:
        basis = both_bases[:, 1]
    else:
        basis = both_bases[:, 0]*0.0    

    if basis[0] < 0:
        basis = -basis

    return basis
```

```python
def bayes_estimator(mu0, var0, var_l, vals):
    mu = np.zeros(vals.shape[0])
    var = np.zeros(vals.shape[0])

    sum_vals = vals[0]
    var[0] = (var_l*var0)/(var_l+var0)
    mu[0] = var[0]*(mu0/var0 + sum_vals/var_l)

    for i in range(1, vals.shape[0]):
        sum_vals += vals[i]
        var[i] = 1/((i+1)/var_l + 1/var0)
        mu[i] = var[i]*(mu0/var0 + sum_vals/var_l)

    return mu, np.sqrt(var)
```

# Minimum Variance Estimator:

```python
def mve_estimate(data, natural_handicap, windows=[100, 200, 300], half_lives=[10, 20, 50, 100]):
    K = len(windows) + len(half_lives)

    mu = np.zeros((data.shape[0], K))
    sigma = np.zeros((data.shape[0], K))

    for i, l in enumerate(windows):
        mu_ = np.array([get_basis(data[(j-l):j, :])[0] for j in range(l, data.shape[0])])
        mu_, sigma_ = bayes_estimator(0.4986 + natural_handicap*0.0024, (2*0.0634)**2, 0.2**2, mu_)

        mu[:l, i] = 0.0
        mu[l:, i] = mu_

        sigma[:l, i] = np.inf
        sigma[l:, i] = sigma_

    for i, hl in enumerate(half_lives):
        mu_ = np.array([get_basis(data[:j, :])[0] for j in range(data.shape[0])])
        mu[:, K-i-1] = pd.ewma(mu_, halflife=hl)
        sigma[:, K-i-1] = pd.ewmstd(mu_, halflife=hl)

    mu[np.isnan(mu)] = 0.0
    sigma[np.isnan(sigma) | (sigma == 0.0)] = np.inf

    var = np.power(sigma, 2)
    var_inv = 1/var
    var_inv_sum = np.sum(var_inv, axis=1)

    weights = var_inv/var_inv_sum.reshape(-1, 1)

    estimator_mu = np.sum(weights*mu, axis=1)
    estimator_var = 1/var_inv_sum
    estimator_sigma = np.sqrt(estimator_var)

    return weights, estimator_mu, estimator_sigma
```

```python
def plot_estimator(eid):
    data = PBP.loc[eid, ['points_h', 'points_a']].values
    line = NH.loc[eid].ftps_1_line
    x = np.arange(data.shape[0])
    actual_basis = get_basis(data)

    weights, estimator_mu, estimator_sigma = mve_estimate(data, line)

    f, (ax1, ax2) = plt.subplots(2, figsize=(16, 12), sharex=True)
    ax1.set_xlim([0, x[-1]])
    ax1.set_ylim([-1.0, 1.0])

    ax1.axhline(actual_basis[0], color='k')
    ax1.plot(x, estimator_mu)
    ax1.fill_between(x, estimator_mu-3*estimator_sigma, estimator_mu+3*estimator_sigma, alpha=0.1)

    ax1.axhline(actual_basis[1], color='k')
    ax1.plot(x, estimator_mu-1)
    ax1.fill_between(x, estimator_mu-3*estimator_sigma-1, estimator_mu+3*estimator_sigma-1, alpha=0.1)

    ax1.set_title('Combined estimator $\\mu$ with 3$\\sigma$ CI')

    ax2.step(x, weights)
    ax2.set_title('Estimator weights')

    plt.tight_layout()
```

```python
plot_estimator(2352975)
```

```python
def get_estimate_errors(eid):
    data = PBP.loc[eid, ['points_h', 'points_a']].values
    line = NH.loc[eid].ftps_1_line
    actual_basis = np.ones(data.shape[0])*get_basis(data)[0]

    _, mu, sigma = mve_estimate(data, line)

    return np.array([
        np.power(actual_basis-mu, 2),
        (actual_basis-mu)/sigma
    ]).T
```

```python
def mse(n):
    from tqdm import tqdm

    errors_sum = np.zeros((400, 2))
    errors_sumsq = np.zeros((400, 2))

    for eid in tqdm(eids[:n]):
        errors = get_estimate_errors(eid)[:400, :]

        errors_sum += errors
        errors_sumsq += np.power(errors, 2)

    error_mean = (errors_sum / n)[:, 0]
    error_std = (errors_sumsq / n - np.power((errors_sum / n), 2))[:, 0]

    return error_mean, error_std
```

```python
mse_mean, mse_std = mse(len(eids))

plt.plot(np.arange(400), mse_mean)
plt.fill_between(np.arange(400), mse_mean-3*mse_std, mse_mean+3*mse_std, alpha=0.1)
```
