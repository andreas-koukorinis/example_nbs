# CO method

The plots are as follows:

1) Potential return for each of the constituent bets if cashed-out at the x-value odds.

2) Evolution of the expected return of the whole portfolio.

3) Surface plot of the return surface - what the portfolio would make if cashed-out at the x-y odds. Note that the matched odds are given by the black dot, the closed odds by the cross and the arrow represents where the signal told us it would go.
